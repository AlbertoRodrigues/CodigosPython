```python
import pandas as pd
from plotnine import *
from sklearn.datasets import load_iris
import seaborn as sea
```

```python
dados=pd.DataFrame(load_iris().data,columns=load_iris().feature_names)
dados.head()
```

```python
x=[]
for i in load_iris().target:
    if i==0:
        x.append("setosa")
    elif i==1:
        x.append("versicolor")
    else:
        x.append("virginica")
dados["Resposta"]=x
```

```python
(ggplot(dados)+ aes("sepal length (cm)","sepal width (cm)",color="Resposta")
+ geom_point()+ggtitle("Gráfico de pontos separado pelo atributo alvo"))
```

```python

(ggplot(dados)+aes("sepal length (cm)")
+geom_histogram(fill="blue",bins=15,color="black")
+ggtitle("Histograma do tamanho da sépala!")+ylab(" "))
```

```python

(ggplot(dados)+aes("sepal length (cm)",fill="Resposta")
+geom_histogram(bins=15,color="black")
+ggtitle("Histograma do tamanho da sépala separado pela variável alvo")+facet_wrap("~Resposta")+geom_rug())

```

```python
(ggplot(dados)+aes("sepal length (cm)")
+geom_histogram(aes(y ="..density.."),fill="blue",bins=15,color="black")
+geom_density()
+ggtitle("Histograma da variável tamanho da sépala e estimativa da função de densidade de probabilidade por Kernel"))

```

```python
(ggplot(dados)+ aes("sepal length (cm)","sepal width (cm)",shape="Resposta",color="Resposta")
+ geom_point())

```

```python
(ggplot(dados)+aes(1,"sepal length (cm)")+geom_boxplot(fill="blue")
+ylab(" ")+xlab("Tamanho da sépala ")+ggtitle("Boxplot da variável tamanho da sépala"))

```

```python
(ggplot(dados)+aes("Resposta","sepal length (cm)",fill="Resposta")
+xlab("Tamanho da sépala ")+ylab(" ")+ggtitle("Box-plots")
+labs(fill="Espécies")+geom_boxplot())

```

```python


(ggplot(dados)+aes("Resposta")+geom_bar(fill="blue",color="black")+xlab("Espécies")
+ geom_text(
     aes(label='stat(count)'),
     stat='count',
     nudge_x=-0.14,
     nudge_y=0.125,
     va='bottom'
 )
 + geom_text(
     aes(label='stat(prop)*100', group=1),
     stat='count',
     nudge_x=0.14,
     nudge_y=0.125,
     va='bottom',
     format_string='({:.1f}%)'
 ))
```
